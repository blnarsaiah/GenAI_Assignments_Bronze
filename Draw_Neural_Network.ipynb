```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyArrowPatch, Rectangle
```

```python
def neuron(ax, x, y, r=0.22, fc="#ffffff", ec="#333333"):
    c = Circle((x, y), r, fc=fc, ec=ec, lw=1.5)
    ax.add_patch(c)

x_input, x_hidden, x_out = 0.5, 2.5, 4.0

y_inputs  = [3.0, 2.4, 1.8, 1.2, 0.6, 0.0]
y_hidden  = [2.4, 1.6, 0.8, 0.0]
y_output  = [1.2]

fig, ax = plt.subplots(figsize=(9, 5))
ax.set_xlim(0, 4.8); ax.set_ylim(-0.5, 3.4)
ax.axis("off")

ax.text(x_input, 3.25, "INPUTS (6)", ha="center", fontsize=11, weight="bold")
ax.text(x_hidden, 3.25, "HIDDEN (4)", ha="center", fontsize=11, weight="bold")
ax.text(x_out,   3.25, "OUTPUT (1)", ha="center", fontsize=11, weight="bold")
ax.set_title("Architecture of Neural Network", fontsize=14)
labels = [
    "TransactionAmount",
    "TransactionTime",
    "MerchantCategory",
    "CustomerAge",
    "AccountBalance",
    "NumberOfTransactionsToday"
]
for y, lab in zip(y_inputs, labels):
    neuron(ax, x_input, y)
    ax.text(x_input - 0.05, y, "  " + lab, va="center", ha="right", fontsize=9)

for y in y_hidden:
    neuron(ax, x_hidden, y, fc="#eef5ff")

neuron(ax, x_out, y_output[0], fc="#fff5e6")
ax.text(x_out + 0.45, y_output[0],
        "Fraud\n(0 = Genuine, 1 = Fraud)", va="center", fontsize=9)

for yi in y_inputs:
    for yh in y_hidden:
        ax.add_patch(FancyArrowPatch((x_input+0.20, yi), (x_hidden-0.20, yh),
                                     arrowstyle='-|>', mutation_scale=8,
                                     lw=0.7, color="#600"))

for yh in y_hidden:
    ax.add_patch(FancyArrowPatch((x_hidden+0.20, yh), (x_out-0.20, y_output[0]),
                                 arrowstyle='-|>', mutation_scale=9,
                                 lw=0.9, color="#400"))



plt.show()
```
