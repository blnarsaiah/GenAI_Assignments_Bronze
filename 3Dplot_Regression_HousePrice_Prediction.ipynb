```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
Area = np.array([600, 750, 900, 1200, 1500, 1800]).reshape(-1, 1)
Bedrooms = np.array([1, 2, 2, 3, 3, 4]).reshape(-1, 1)
Price_Lakhs = np.array([25, 35, 45, 60, 80, 95])

X = np.column_stack((Area, Bedrooms))

model = LinearRegression()
model.fit(X, Price_Lakhs)

area_grid = np.linspace(Area.min(), Area.max(), 25)
bed_grid = np.linspace(Bedrooms.min(), Bedrooms.max(), 25)
A, B = np.meshgrid(area_grid, bed_grid)

X_grid = np.column_stack((A.ravel(), B.ravel()))
Z = model.predict(X_grid).reshape(A.shape)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Area, Bedrooms, Price_Lakhs, color='blue', s=45)
ax.plot_surface(A, B, Z, alpha=0.4, color='orange')

ax.set_title("3D House Price Prediction (in Lakhs)")
ax.set_xlabel("Area (sq ft)")
ax.set_ylabel("Number of Bedrooms")
ax.set_zlabel("Price (Lakhs ₹)")

plt.show()
```
